```python
# Loading Libraries
import pandas as pd
import numpy as np
```

```python
# Reading csv Files
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train.info()
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
df_train = df_train.drop(['PassengerId','Name','Ticket','Cabin'], axis=1)
df_test = df_test.drop(['PassengerId','Name','Ticket','Cabin'], axis=1)
```

```python
print(f"The number of null in training set are \n{df_train.isnull().sum()}")
```

```python
print(f"The number of null in testing set are \n{df_test.isnull().sum()}")
```

```python
df_train['Age'] = df_train['Age'].fillna(df_train['Age'].mean())
df_test['Age'] = df_test['Age'].fillna(df_test['Age'].mean())
df_train['Embarked'] = df_train['Embarked'].fillna(df_train['Embarked'].mode()[0])
df_test['Fare'] = df_test['Fare'].fillna(df_train['Fare'].mean())
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
# Encoding the columns into One-Hot Encoding
from sklearn.preprocessing import OneHotEncoder
columns_to_encode = ['Pclass','Sex','Embarked']
ohe = OneHotEncoder()
encoded_train = ohe.fit_transform(df_train[columns_to_encode]).toarray()
encoded_test = ohe.transform(df_test[columns_to_encode]).toarray()
```

```python
X_train = df_train.drop(columns_to_encode, axis=1)
X_test = df_test.drop(columns_to_encode, axis=1)
```

```python
X_train[:10]
```

```python
# Separating the target variable
y_train = X_train['Survived'].values
X_train = X_train.drop('Survived', axis=1).values
X_test = X_test.values
```

```python
X_train = np.concatenate((X_train, encoded_train), axis=1)
X_test = np.concatenate((X_test, encoded_test), axis=1)
```

```python
X_train[:10]
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,[0,3]] = sc.fit_transform(X_train[:,[0,3]])
X_test[:,[0,3]] = sc.transform(X_test[:,[0,3]])
```

```python
# Training the model on random forest
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=42)
classifier.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
accuracy = cross_val_score(classifier,X_train,y_train,cv=10,scoring='accuracy').mean()
print(f"The accuracy of the model is {accuracy}")
```

```python
# Training the model on Support Vector Machine
from sklearn.svm import SVC
classifier = SVC(kernel='rbf', random_state=42)
classifier.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
accuracy = cross_val_score(classifier,X_train,y_train,cv=10,scoring='accuracy').mean()
print(f"The accuracy of the model is {accuracy}")
```
