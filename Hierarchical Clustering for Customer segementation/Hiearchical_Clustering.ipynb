```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading the dataset
df = pd.read_csv('Mall_Customers.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,[3,4]].values
```

```python
# Using dendogram to find cluster
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

As, for finding the number of cluster we have to find the vertical line with highest distance without cutting by horizontal line. Look from above and find highest vertical line without cutting by horizontal line. As we see in dendogram, left blue line and orange right side line are comparatively equal. so we can find the number of cluster by cutting the line in that lines. By cutting blue line, we got three clusters and by cutting orange we get five cluster.

```python
# Training dataset using Agglomertaive clustering using n_clusters = 3
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='ward')
cluster.fit(X)
y = cluster.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Clustering using hiearchical clustering')
plt.show()
```

As, we used three cluster which not effective clutser because the Annual income from 20 to 40 cluster as same cluster which can be differentitate based on spending score. so use five cluster

```python
# Training dataset using Agglomertaive clustering using n_clusters = 5
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='ward')
cluster.fit(X)
y = cluster.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Clustering using hiearchical clustering')
plt.show()
```
