```python
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading dataset
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
plt.hist(df['Age'])
plt.show()
```

```python
plt.hist(df['Genre'])
plt.show()
```

```python
df.var(numeric_only=True)
```

As we have serial wise customer id so we ignore it.

```python
# Elbow method to find the optimal cluster
X = df.iloc[:,[3,4]].values
from sklearn.cluster import KMeans
wccs = []
for i in range(1,11):
  cluster = KMeans(n_clusters=i,init='k-means++',random_state=42)
  cluster.fit(X)
  wccs.append(cluster.inertia_)
```

```python
plt.plot(range(1,11),wccs)
plt.xlabel('Number of Clutsers')
plt.ylabel('WCCS')
plt.title('Elbow Method')
plt.show()
```

As from elbow metjod, we can find that the number of cluster is 5 as the line after 5 goes smoothly.

```python
# Training dataset on K means
cluster = KMeans(n_clusters=5,init='k-means++',random_state=42)
cluster.fit(X)
y = cluster.labels_
```

```python
# scatter plot with colors based on cluster labels
scatter = plt.scatter(X[:,0],X[:,1],c = y,cmap = 'viridis',marker='o')
# Lables for axes
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
# title of plot
plt.title('Clustering with K means')
# show the plot
plt.show()
```
